Looking at the purchases, it seems that a purchase of 2000 is much more often than it seems like it should be. Investigating these purchases...

It seems that this purchaser is at least one of the reasons why the AOV is so high! These outliers which do not represent an average order like the definition intended to cover increase everything! 

Upon further thought, I also want to investigate what are the highest purchases of the dataset. A quick and dirty way (because this method does not distinguish between a $2000 order of two items where one shoe is $1 and the other is $1999, and two $1000 shoes) of finding if there are any shoes that are absurdly expensive is simply looking at the highest (order_amounts / total_items)

```python
data['average_per_item'] = data.apply(lambda row: row.order_amount / row.total_items, axis = 1)
```

```python
expensive_items = data.sort_values(by=['average_per_item'], ascending = False)
expensive_items.head(10)
```

Interesting! So there are shoes that have sold for over $25000, multiple times! These shoes also seem to belong to the same shop, and different people are ordering it. This makes it seeme like it's not a mistake (as people are buying it) which makes me wonder if something illegal is going on? This item and shop seems like a case to flag. For the moment, it seems like they are buying not just shoes, because a quick Google search tells me there are not mass-produced shoe that sell for this much. 

In conclusion, this calculation is going wrong because:
1. Impossible shoe sales. This shop should be flagged and investigated.
2. Bots buying shoes, which should not be counted for the AOV metric.
